## Genomes

```python
genome = Genome(1, n_inputs=2, n_outputs=1)
print(genome)
genome.visualize()
```

```python
for _ in range(15):
    genome.mutate()

print(genome)
genome.visualize()
```

```python
parent1 = Genome(1, n_inputs=2, n_outputs=1)
parent2 = Genome(2, n_inputs=2, n_outputs=1)

for i in range(15):
    parent1.mutate()
    parent2.mutate()

parent1.fitness = 1
parent2.fitness = 0

print('Genome1:')
parent1.visualize()

print('Genome2:')
parent2.visualize()

child = Genome(1, n_inputs=2, n_outputs=1)
child.configure_crossover(parent1, parent2)
print('Child:')
child.visualize()
print(child)
```

```python
genome = Genome(1, n_inputs=2, n_outputs=1)
for _ in range(20):
    genome.mutate()

genome.visualize(show_weights=True)
```

```python
network = FeedForwardNetwork.create(genome)
print(network.forward([1, 1]))
network.visualize(genome)
```

```python
genome1 = Genome(1, n_inputs=2, n_outputs=2)
genome2 = Genome(2, n_inputs=2, n_outputs=2)

print(genome1.distance(genome2))

for _ in range(10):
    genome1.mutate()
    genome2.mutate()

genome1.visualize()
genome2.visualize()

print(genome1.distance(genome2))
```

Ab einer Distanz von 3.0 gehrt ein Individuum zu einer neuen Spezies.
